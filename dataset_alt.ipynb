```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data\proposicoes_2013-2024_final_pres_cam.csv')
df
```

```python
cols = [col for col in df.columns if col != 'VirouLei' and col != 'Partido_Pres_Cam'] + ['VirouLei'] + ['Partido_Pres_Cam']
df = df[cols]
df
```

- Destacando 2024

```python
df_2024 = df[df['Ano'] == 2024]
df = df[df['Ano'] != 2024]
```

- Salvando

```python
df
```

```python
df_2024.reset_index(drop=True, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv('data/proposicoes_ord_2013-2023.csv', index=False)
print("Final DataFrame saved to data/proposicoes_ord_2013-2023.csv")

df_2024.to_csv('data/proposicoes_2024.csv', index=False)
print("Final DataFrame saved to data/proposicoes2024.csv")
```
